# Un example de Notebook

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

N=4096
```

Bon, importations.
Remarque que on peut ecrire du latex: $$\sum_0^ne^{i\pi} = -n$$

```python
H=np.hanning(N)
plt.plot(H)
```

Ca c'est la fenetre de Hanning

```python
S=np.sin(np.arange(N)/30)
plt.plot(S)
```

Bon, c'est un sinus normal. Mtn on va multiplier a par la fenetre.

```python
HS=np.multiply(H,S)
plt.plot(HS,color="red")
```

Maintenant faisons Fourrier.

```python
FS=np.absolute(np.fft.fft(S))
plt.plot(FS[:40])
```

Pas trop de surprises

```python
FHS=np.absolute(np.fft.fft(HS))
plt.plot(FHS[:40])
```
